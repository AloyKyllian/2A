```python
"""
module pour l'état de l'etape 2 ds le cas 2
"""
from projet.outils.GrapheDeLieux import GrapheDeLieux
from projet.etape2.Etat import Etat
from projet.solvers.SolverAStar import SolverAStar

class EtatCas2(Etat) :
    """
    Classe pour definir un etat pour le cas 2 de la tache 2
    """

    # attributs
    # A COMPLETER
    # //////////////////////////////////////////////
    tg : GrapheDeLieux
    lieux_visites=[]
    """ le graphe representant le monde """

    # constructeurs
    # A ECRIRE/MODIFIER/COMPLETER
    # //////////////////////////////////////////////
    def __init__(self, tg : GrapheDeLieux, lieuxDepartVoulu=None, lieuxDejaVisite=None) :
        """ constructeur d'un etat a partir du graphe representant le monde

        :param tg: graphe representant le monde

        :param param1: a definir eventuellement

        :param param2: a definir eventuellement
        """
        self.tg = tg
        
        # a completer pour tenir compte de la presence ou pas des deux derniers parametres
        if lieuxDepartVoulu is not None :
            self.lieuxDepart = lieuxDepartVoulu
            if lieuxDejaVisite is not None:
              self.lieux_visites = list(lieuxDejaVisite)
        else:
            self.lieuxDepart = 0




    # methodes issues de Etat
    # //////////////////////////////////////////////
    def estSolution(self) :
        """ methode detectant si l'etat est une solution

        :return true si l'etat courant est une solution, false sinon
        """
        print(self.__str__())
        # A ECRIRE et MODIFIER le return en consequence and self.lieuxCourant == 0 self.lieuxCourant=self.lieuxFinal en sachant que final = lieuxvistee[-1]
        return len(self.lieux_visites) == self.tg.getNbSommets() and self.lieux_visites[-1]==0

    def successeurs(self):
        """Méthode permettant de récupérer la liste des états successeurs de l'état courant.

        :return Liste des états successeurs de l'état courant.
        """
        successeurs = []
        lieuDepart=self.lieuxDepart
        voisins = self.tg.getAdjacents(self.lieuxDepart)
        for voisin in voisins:
            if voisin not in self.lieux_visites:
                nouveaux_visites = list(self.lieux_visites)
                nouveaux_visites.append(voisin)
                successeurs.append(EtatCas2(self.tg, voisin, nouveaux_visites))
                
        return successeurs



    def h(self) :
        """ methode permettant de recuperer l'heuristique de l'etat courant

        :return heuristique de l'etat courant
        """
        # A ECRIRE et MODIFIER le return en consequence
        distance=0
        for i in range (len(self.lieux_visites)-1):
          if len(self.lieux_visites)>1:
            distance+=GrapheDeLieux.dist(self.lieux_visites[i],self.lieux_visites[i+1],self.tg)
        return distance


    def k(self, e) :
        """ methode permettant de recuperer le cout du passage de l'etat courant à l'etat e

        :param e: un etat

        :return cout du passage de l'etat courant à l'etat e
        """
        # A ECRIRE et MODIFIER le return en consequence
        return self.tg.getCoutArete(self.lieuxDepart,e.lieuxDepart)


    def displayPath(self, pere) :
        """ methode pour afficher le chemin qui a mene a l'etat courant en utilisant la map des peres

        :param pere: map donnant pour chaque etat, son pere
        """
        # A ECRIRE
        chemin = []
        etat_courant = self
        while etat_courant is not None:
            chemin.append(etat_courant.lieuxDepart)
            etat_courant = pere.get(etat_courant, None)
        # chemin.reverse()
        print("resultat trouve : ", chemin)




    # methodes pour pouvoir utiliser cet objet//////
    def __hash__(self) :
        """ methode permettant de recuperer le code de hachage de l'etat courant
        pour une utilisation dans des tables de hachage

        :return code de hachage
        """
        # A ECRIRE et MODIFIER le return en consequence
        return hash(tuple(self.lieux_visites))


    def __eq__(self, o) :
        """ methode de comparaison de l'etat courant avec l'objet o

        :param o: l'objet avec lequel on compare

        :return true si l'etat courant et o sont egaux, false sinon
        """
        # A ECRIRE et MODIFIER le return en consequence
        if isinstance(o, EtatCas2):
          return self.lieuxDepart == o.lieuxDepart
        return False



    # methode pour affichage futur (heritee d'Object)
    # //////////////////////////////////////////////
    def __str__(self) :
        """ methode mettant l'etat courant sous la forme d'une
        chaine de caracteres en prevision d'un futur affichage

        :return representation de l'etat courant sour la forme d'une
        chaine de caracteres
        """
        # A ECRIRE et MODIFIER le return en consequence
        return f"État courant : {self.lieuxDepart}, État visité : {self.lieux_visites}"

# ///////////////////////////////////////////////
# tests sur CAS 2 : tour complet par voie de terre
# ///////////////////////////////////////////
# cas : 10 villes de 0 à 9
tg = GrapheDeLieux.loadGraph("Data/town10.txt",True)
cas2 = EtatCas2(tg)
print("======== TEST CAS 2 10 villes de 0 a 9 : \n")
print(cas2.__str__())
# cas2.liste_ville_deja_visite.clear()
SolverAStar.aStar(cas2)
# choisir ici un algo et l'executer
```
